## Extract Coordinates

```python
# Functions
# Clean Locations String
def extract(a):
    import re
    if type(a)==str and a!="":
        a = a.lower()
        a = re.sub(r'\d+', '',a) # hapus digit
        a = re.sub(r'[\W\s]', ' ',a) # hapus tanda baca
        a = re.sub(' +',' ', a.strip()) # hapus spasi berlebih
        a = ' '.join([i if len(i)!=1 else '' for i in a.split()]).strip() # hapus huruf satu digit (mis: A B C, dsb).
        # Hapus kata
        l = ['wilayah','jalan','jl.','jl','jalur','raya','umum','desa','kelurahan','kel.','kel','nasional',
             'kecamatan','kec.','kec','kabupaten','kab.','kab','kota','ruas','dusun','jurusan',
            'tepatnya','tersebut','pertigaan','perempatan','simpang','tiga','empat', 'pos',
             'rt','rt.','rw','rw.','km','km.']
        ai = a.split()
        out = []
        i = 0
    else:
        return '';
    
    while i < len(ai):
        temp = []
        while i < len(ai):
            if ai[i] not in l:
                temp.append(ai[i])
            else: break;
            i+=1
        if len(temp)!=0:
            out.append(' '.join(temp))
        i+=1
    out = ','.join(out)
    return out

# Get coordinates
locator = Nominatim(user_agent='myGeocoder')
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)

def add(g):
    idx = g[0].split(', ').index('Jawa Timur')
    loc=None;prov=None; kabkot=None; desakel=None; jalan=None
    loc = g[0]
    point = g[1]
    prov = g[0].split(', ')[(idx)] # get province
    kabkot = g[0].split(', ')[(idx-1)] # get kab/kot
    if idx-1>0: desakel = g[0].split(', ')[(idx-2)] # get desa/kel
    if idx-3>=0: jalan = ' '.join(g[0].split(', ')[:(idx-2)]) # get nama jalan
    return [loc,point,jalan,desakel,kabkot,prov,g[1][0],g[1][1]]

def getloc(e):
    none = [None,None,None,None,None,None,None,None]
    #Bila ada alamat
    if e!='':
        g = geocode(e)
        if g!=None:
            if 'Jawa Timur' not in g[0].split(', '):
                e+=',jawa timur'
                g = geocode(e)
                if g!=None: 
                    if 'Jawa Timur' in g[0].split(', '): return add(g);
                    else: return none
                else: return none;
            else: 
                if 'Jawa Timur' in g[0].split(', '): return add(g)
                else: return none
        else:
            e+=',jawa timur'
            g = geocode(e)
            if g!=None:
                if 'Jawa Timur' in g[0].split(', '): return add(g);
                else: return none;
            else: return none;
    else:
        return none;
```

## Extract Time

```python
# TIMES
# Extract times

import re
from datetime import datetime

def is_time_format(s):
    time_re = re.compile(r'^(([01]\d|2[0-3]).([0-5]\d)|00.00)$')
    return bool(time_re.match(s))

art_idx = pd.unique(data.article)
progbar = Progbar(len(art_idx))
times = [];

for idx in range(len(art_idx)):
    art_df = data[data.article==art_idx[idx]]
    sent_id = pd.unique(art_df.sentence)
    time_array = []; #untuk menampung time di tiap kalimat
    # sentence
    for i in range(len(sent_id)):
        sent_df = art_df[art_df.sentence==sent_id[i]]
        time_str = "";
        FOUND=False     
        # word
        for j in range(len(sent_df)):
            token = sent_df.iloc[j]
            if "TIME" in token.pred_ner:
                if is_time_format(token.word):
                    try:
                        d = datetime.strptime(token.word,"%H.%M")
                    except:
                        continue
                    if datetime.strptime("00.00","%H.%M")<=d<datetime.strptime("12.00","%H.%M"):time_str="pagi";FOUND=True;break;
                    elif datetime.strptime("12.00","%H.%M")<=d<datetime.strptime("15.00","%H.%M"):time_str="siang";FOUND=True;break;
                    elif datetime.strptime("15.00","%H.%M")<=d<datetime.strptime("18.00","%H.%M"):time_str="sore";FOUND=True;break;
                    elif datetime.strptime("18.00","%H.%M")<=d<=datetime.strptime("23.59","%H.%M"):time_str="malam";FOUND=True;break;
                if not FOUND:
                    for a in ['pagi','siang','sore','malam']:
                        if token.word.lower() in a: time_str = a; FOUND=True;break;
                    if token.word.lower()=='petang':time_str='malam';FOUND=True;break;
                    if 'dini' in token.word.lower():time_str='pagi';FOUND=True;break;
        if FOUND:
            times.append(time_str)
            break;
    if not FOUND:
        times.append(None)
    progbar.update(idx+1);

# Print some output time
times[:10]
```
