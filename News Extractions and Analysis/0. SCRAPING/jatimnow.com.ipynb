```python
# set path
path = r"C:\Users\Dwi Nissa\AppData\Local\Programs\Python\Python38\Lib\site-packages\selenium\webdriver\firefox\geckodriver.exe"
options = webdriver.FirefoxOptions()
options.set_preference("dom.push.enabled", False)
driver = webdriver.Firefox(executable_path = path, firefox_options=options)

news = []; page_begin = 1; page_to_scrap = 20
print("Scrap {} item".format((page_to_scrap-page_begin+1)*14))

# Scraping informasi berita online
for p in range(page_begin,page_to_scrap+1):
    url = "https://jatimnow.com/search/tabrak/page/"+str(p)
    r = driver.get(url)
    sleep(2)
    boundaries = [tag.find_elements_by_css_selector('.punica-zoom-effect') for tag in driver.find_elements_by_css_selector('.col-a')][0]
    links = [art.find_element_by_tag_name('a').get_attribute('href') for art in boundaries]
    categories = [cat.text for cat in driver.find_elements_by_css_selector('.entry-categories')]
    if len(links)==len(categories):
        entry_cat = True
    for i in range(len(links)):
        content = ''
        if entry_cat: category = categories[i]
        request = driver.get(links[i])
        try:
            datetimes = []
            tag_datetime = [tag.find_elements_by_css_selector('.entry-date') for tag in driver.find_elements_by_css_selector('.entry-box')]
            for dt in tag_datetime[0]:
                datetimes.append(dt.text)
            datetime = ' '.join(datetimes)
        except(NoSuchElementException, IndexError, ValueError): pass
        try:
            title = driver.find_element_by_css_selector('.entry-title').text
        except(NoSuchElementException, IndexError, ValueError): pass
        try:
            contents = []
            tag_content = driver.find_elements_by_css_selector("div[itemprop='articleBody']")
            for c in tag_content:
                contents.append(c.text)
            content = ''.join(contents)    
        except(NoSuchElementException, IndexError, ValueError): pass
        news.append([links[i],datetime,title,content,category])
print('DONE.')
if len(news)==(page_to_scrap-page_begin+1)*14:
    print('All news scraped')
else:
    print('Berita yg ga ter scrap:{}'.format((page_to_scrap-page_begin+1)*14-len(news)))
print('TIME:{} minutes'.format((time()-start_time)/60))
```
