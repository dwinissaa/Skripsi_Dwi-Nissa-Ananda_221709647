## Import Libraries

```python
import pandas as pd 
import numpy as np
from ast import literal_eval
```

```python
data = pd.read_csv("./../5. REMOVE DUPLICATION/OUTPUT/FILTERED_EXTRACTED_FINAL.csv", index_col=0)
# Read Umur as Array
data['age'] = data['age'].apply(lambda x: literal_eval(x) if type(x)==str else None)
data
```

## Umur

```python
# Define range
num_range = []
num = list(range(0,61,5))
n = 0;
while n < len(num)-1:
    num_range.append(range(num[n],num[n+1]))
    n+=1;
# Define Labels
lab = [f"{num[i]}-{num[i+1]-1}" for i in range(len(num_range))]
lab.append(f">{num[len(num_range)]}")

# Categorize ages
agec = []
for row in data.age:
    li_age = []
    if type(row)==list:
        for i in row:
            for ni, nr in enumerate(num_range):
                if i in nr: li_age.append(lab[ni]); break
                else:
                    if i>=60: li_age.append(lab[len(num_range)]); break
        agec.append(li_age)
    else:
        agec.append(None)
        
# Define Labels
lab = [f"{num[i]}-{num[i+1]-1}" for i in range(len(num_range))]
lab.append(f">{num[len(num_range)]}")
age = {}
for i in lab:
    age[i] = 0
for i in agec:
    if type(i)==list:
        for j in i:
            age[j] += 1
df_umur = pd.DataFrame({'count':[]})
for i in range(len(list(age))):
    df_umur.loc[list(age)[i],'count'] = age[list(age)[i]]
df_umur.index.name = 'umur'
df_umur
```

```python
df_umur.to_csv('./OUTPUT/UMUR.csv')
```

## Jumlah Berita menurut Bulan, Hari

```python
from datetime import datetime

datetime.strptime('2020-12-30', '%Y-%m-%d')
dates_format = data['date'].apply(lambda x: datetime.strptime(x, '%Y-%m-%d') if type(x)==str else None)
```

```python
df_bulan = pd.DataFrame(dates_format.groupby(dates_format.dt.to_period("M")).agg('count'))
df_bulan.index.name = 'bulan'
df_bulan.columns=['kejadian']
df_bulan
```

```python
df_hari = pd.DataFrame(dates_format.groupby(dates_format.dt.strftime("%A")).agg('count'))
df_hari.index.name = 'hari'
df_hari.columns=['kejadian']
df_hari_ =pd.DataFrame()
hari_dict = {'Monday':'Senin',
             'Tuesday':'Selasa',
             'Wednesday':'Rabu',
             'Thursday':'Kamis',
             'Friday':'Jumat',
             'Saturday':'Sabtu',
             'Sunday':'Minggu',}
for i in list(hari_dict):
    df_hari_.loc[hari_dict[i],'kejadian'] = int(df_hari.loc[i,'kejadian'])
df_hari_.index.name = 'hari'
df_hari_
```

```python
df_hari_.to_csv("./OUTPUT/HARI.csv")
df_bulan.to_csv("./OUTPUT/BULAN.csv")
```

## Peta Titik Lokasi Kecelakaan

Hanya data yang punya jalan saja yang diekspor ke GeoJSON

```python
data_geo = data.dropna(subset=['jalan','date'])
data_geo
```

```python
import csv, json
from geojson import Feature, FeatureCollection, Point

features = []

for i in range(len(data_geo)):
    latitude = data_geo.iloc[i,6];
    longitude = data_geo.iloc[i,7];
    location = data_geo.iloc[i,4];
    date = data_geo.iloc[i,12];
    features.append(
        Feature(
            geometry = Point((latitude,longitude)),
            properties = {
                'date': date,
                'location':location
            }
        )
    )
collection = FeatureCollection(features)
name = "GeoObs_FINAL"
with open(f"./OUTPUT/{name}.json", "w") as f:
    f.write('%s' % collection)
    print(f"JSON written in: ./OUTPUT/{name}.json")
```
