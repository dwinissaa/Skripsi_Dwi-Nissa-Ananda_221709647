## Import Libraries

```python
import pandas as pd
import numpy as np
import math
from ast import literal_eval
from sklearn.preprocessing import MultiLabelBinarizer
```

```python
data = pd.read_csv("./../4. RULE BASED/OUTPUT/EXTRACTED_FINAL.csv",index_col=0)
data
```

## Filter Data

```python
# Ambil column yg diperlukan ajaa
filtered = pd.DataFrame()
for i in ['date','time','loc']:
    filtered[i] = data[i]
# Bila ditemukan kesamaan: lokasi,date,time nya-> remove duplicates, keep='last'
drop_na = filtered.sort_values(by=['date','time','loc'],ascending=False).dropna(subset=['loc','date','time']).index
na_idx = [i for i in filtered.index if i not in drop_na]
uniq_idx = pd.concat([filtered.loc[drop_na].drop_duplicates(subset=['date','time','loc'],keep='last'),
                    filtered.loc[na_idx]]).index
data_filtered = data.loc[uniq_idx]
data_filtered
```

```python
data_filtered.to_csv("./OUTPUT/FILTERED_EXTRACTED_FINAL.csv")
```

## Data untuk Membuat Decision Tree di RStudio

```python
data = pd.read_csv("./OUTPUT/FILTERED_EXTRACTED_FINAL.csv",index_col=0)
# Read Umur as Array
data['age'] = data['age'].apply(lambda x: literal_eval(x) if type(x)==str else None)
data['target']= [1 if data.loc[i,'die']>0 else 0 for i in data.index]
#[data.loc[i,'die']+data.loc[i,'injured'] for i in data.index]
analyze = data[['day','time','age','truck','car','motorcycle','bicycle','train','bus','die','injured','target']]
# drop na
analyze = analyze.dropna(subset=['age','time','day'])
analyze.columns=['day','time','age','involve_truck','involve_car','involve_motorcycle','involve_bicycle','involve_train','involve_bus','die','injured','target']
analyze
```

### Categorize Age

```python
# Define range
num_range = []
num = list(range(0,61,5))
n = 0;
while n < len(num)-1:
    num_range.append(range(num[n],num[n+1]))
    n+=1;

# Define Labels
lab = [f"{num[i]}-{num[i+1]-1}" for i in range(len(num_range))]
lab.append(f">{num[len(num_range)]}")

# Categorize ages
agec = []
for row in analyze.age:
    li_age = []
    for i in row:
        for ni, nr in enumerate(num_range):
            if i in nr: li_age.append(lab[ni]); break
            else:
                if i>=60: li_age.append(lab[len(num_range)]); break
    agec.append(li_age)

# One-Hot Encoding Ages
test = pd.Series(agec)
mlb = MultiLabelBinarizer()
res = pd.DataFrame(mlb.fit_transform(test),
                   columns=[f"age_{i}" for i in mlb.classes_],
                   index=analyze.index)
res
```

### Data siap olah di RStudio:

```python
a = analyze[['day','time']]
for c in res.columns:
    a[c]=res[c]
for d in ['involve_truck','involve_car','involve_motorcycle','involve_bicycle','involve_train','involve_bus']:
    a[d]=analyze[d]
a['target'] = analyze['target']
a
```

Data siap olah di RStudio diekspor ke .csv

```python
a.to_csv("./OUTPUT/decision_tree_data.csv")
```
